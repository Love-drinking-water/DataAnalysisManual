```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# 0是最小值  14是最大值  分为100份
x = np.linspace(0, 14, 100) 
```

```python
y1 = np.sin(x)
y2 = np.sin(x+2)*1.25
```

```python
def sinplot():
    plt.plot(x, y1)
    plt.plot(x, y2)
```

```python
sinplot()
```

```python
# 引入seaborn
import seaborn as sns
```

```python
sinplot()
```

```python
# seaborn的五种风格
style = ['darkgrid', 'dark', 'white', 'whitegrid', 'ticks']
sns.set_style(style[0], {'grid.color': 'red'})
sinplot()
```

## 强大的调色功能

```python
def sinplot_four():
    x = np.linspace(0, 14, 100)
    # 设置图像大小
    plt.figure(figsize=(12, 8))
    for i in range(4):
        plt.plot(x, np.sin(x+i)*(i+0.75), label='sin(x+%s)*(%s+0.75)'%(i, i))
    plt.legend()
```

```python
sinplot_four()
```

```python
# RGB 取值
sns.color_palette()
```

```python
pal_style = ['deep', 'muted', 'pastel', 'bright', 'dark', 'colorblind']
# 画出调色板
sns.palplot(sns.color_palette('dark'))
```

```python
# 重新定义颜色
sns.set_palette(sns.color_palette('dark'))
```

```python
sinplot_four()
```

```python
# 清除设置
sns.set()
```

```python
sinplot_four()
```

```python
# 使用with语句设置颜色 在with中的语句才会使用设置的颜色  with以外的语句使用默认的颜色设置
```

```python
with sns.color_palette('dark'):
    sinplot_four()
```

```python
sinplot_four()
```
