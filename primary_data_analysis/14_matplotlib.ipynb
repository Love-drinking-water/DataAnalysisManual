```python
# 在一个图像里面显示多条线
t = np.arange(0.0, 2.0, 0.1)
t.size
```

```python
s = np.sin(t*np.pi)
```

```python
plt.plot(t, s, 'r--', label='aaa')
plt.plot(t*2, s, '--', label='bbb')
# 设置坐标的label
plt.xlabel('this is x')
plt.ylabel('this is y')
# 设置标题
plt.title('this is title')
# 显示样例
plt.legend()
```
