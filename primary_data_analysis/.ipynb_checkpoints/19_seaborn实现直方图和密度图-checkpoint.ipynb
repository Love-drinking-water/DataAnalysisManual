```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from pandas import Series, DataFrame
%matplotlib inline
```

```python
import seaborn as sns
```

```python
s1 = Series(np.random.randn(1000))
```

```python
# 直方图
plt.hist(s1)
```

```python
# 密度图
s1.plot(kind='kde')
```

```python
# 使用seaborn画图
# s1 传入的数据
# hist 是否画直方图
# kde 是否画密度图
# rug 是否显示密度分布
# bins 分成多少份

sns.distplot(s1, hist=True, kde=True, rug=True)
```

```python
# kdeplot
# shade 是否填充
# vertical 垂直或水平排列


sns.kdeplot(s1, shade=True, color='r')
```
