```python
import numpy as np
import pandas as pd
from pandas import Series
import matplotlib.pyplot as plt
```

```python
# s1 累加和
s1 = Series(np.random.randn(1000)).cumsum()
s2 = Series(np.random.randn(1000)).cumsum()
```

```python
# kind 显示的图像类型
# grid 是否显示方格
# label 标签
# title 标题
# style 显示的线的形状

s1.plot(kind='line', grid=True, label='s1', title='this is a Series')
s2.plot(kind='line', grid=True, label='s2', title='this is a Series')

plt.legend()
```

```python
# 使用subplots生成子图
fig, ax = plt.subplots(2, 1)
ax[0].plot(s1)
ax[1].plot(s2)
```

```python
# 使用series生成子图
fig, ax = plt.subplots(2, 1)
s1[:15].plot(ax=ax[0], label='s1', kind='bar')
s2.plot(ax=ax[1], label='s2')
```

## DataFrame画图

```python
from pandas import Series, DataFrame
```

```python
df = DataFrame(
    np.random.randint(1, 10, 40).reshape(10, 4),
    columns=['A', 'B', 'C', 'D']
)

# kind 显示的图像类型
# grid 是否显示方格
# label 标签
# title 标题
# style 显示的线的形状
# 竖向显示
# df.plot(kind='bar')
# 横向显示
# df.plot(kind='barh')
# 叠加显示
df.plot(kind='bar', stacked=True)
```

```python
# 填充显示
df.plot(kind='area')
```

```python
# 按照行进行画图
a = df.iloc[5]
a.plot()
```

```python
# 按照行画图，画多行
for i in df.index:
    df.iloc[i].plot(label=str(i))
plt.legend()
```

```python
# 显示单列图
df['A'].plot()
```

```python
# 按行画图  可以使用先转制 再画图
df.T.plot()
```
