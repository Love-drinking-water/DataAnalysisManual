```python
import numpy as np
import pandas as pd
from pandas import Series, DataFrame
```

```python
df = pd.read_csv('./data/apply_demo.csv')
```

```python
df.head()
```

```python
df.size
```

```python
# 添加一列
s1 = Series(['a']*7978)
```

```python
df['A'] = s1
```

```python
df.head()
```

```python
# 修改某一列的值
df['A'] = df['A'].apply(str.upper)
```

```python
df.head()
```

```python
# 将其中的data变成三列
def foo(line):
#     根据空格进行切分
    items = line.strip().split(' ')
    return Series([items[1], items[3], items[5]])
```

```python
df_tmp = df['data'].apply(foo)
```

```python
# 重命名列名
df_tmp = df_tmp.rename(columns={0:'Symbol', 1:'Seqno', 2:'Price'})
```

```python
df_tmp
```

```python
df.head()
```

```python
# 合并
df_new = df.combine_first(df_tmp)
```

```python
df_new
```

```python
del df_new['data']
```

```python
del df_new['A']
df_new
```

```python
df_new.to_csv('./data/demo_duplicate_1.csv')
```
